### 2.) Darstellung der Exponentialfunktion:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def createPlot(xAxis ,function, title, xlabel, steps, ylabel):
    plt.plot(xAxis,function)
    plt.grid(True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(np.linspace(xAxis[0],xAxis[-1],steps))
```

```python
n=500
t=np.linspace(-5,6,n)
f=0.4
omega=2*np.pi*f
im=np.sin(omega*t)
re=np.cos(omega*t)
```

```python
plt.figure(figsize=(15,10))
ax = plt.axes(projection='3d')
ax.plot3D(t,im,-1, "green")
ax.plot3D(t,np.ones(n),re, "blue")
ax.plot3D(t,im,re, "red")
ax.set_xlabel('t', size="26", labelpad=15)
ax.set_ylabel('$\mathfrak{I}$($ e^{j\omega t}$)',size="26", labelpad=15)
ax.set_zlabel('$\mathfrak{R}$($ e^{j\omega t}$)',size="26", labelpad=15);
```

### 3.) Numerische Lösung Integral:

```python
n=32
f2=40
omega=2*np.pi*f2
t2=np.linspace(0,2/f2,n)
f_t=np.sin(omega*t2)
F_t=-np.cos(2*np.pi*f2*t2)
plt.subplot(2,1,1)
createPlot(t2,f_t,"Sinus-Signal","t",10,"f(t)")
plt.subplot(2,1,2)
plt.subplots_adjust(hspace=1)
createPlot(t2,F_t,"Stammfunktion","t",10,"F(t)")
plt.title("Stammfunktion:")
```

```python
n2=600
start2=0
end2=4
rect_t=np.zeros(n2)
tao1=1
tao2=2.5
u=2
t3=np.linspace(start2,end2,n2)
for i in range(0,n2):
    if t3[i]>=tao1 and t3[i]<=tao2:
        rect_t[i]=u
plt.subplot(2,1,1)
createPlot(t3,rect_t,"Rechtecksignal:","t",5,"f(t)")
plt.subplot(2,1,2)
plt.subplots_adjust(hspace=1)
createPlot(t3,u*(tao2-tao1)*np.ones(n2),"Integrationsergebnis:","t",6,"F(t)")
 
```

### Berechnungen:

```python
dt=2/len(f_t)
print(dt)
F_k=[]
F_k.append(f_t[0]*dt)
for k in range(1,n):
    F_k.append(F_k[k-1]+(f_t[k-1]*dt))
plt.subplot(2,1,1)
createPlot(t2,F_k,"Stammfuntion numerisch:","t",3,"F(t)")
```

```python
dt2=end2/n2
F_k2=np.zeros(n2)
F_kk2=np.zeros(n2)
for k in range(0,n2):
    F_k2=F_k2+rect_t[k]*dt2
    F_kk2[k]=F_k2[k]
plt.subplot(2,1,1)
plt.subplots_adjust(hspace=1)
createPlot(t3,F_k2,"Integrationsergebnis Rechteck:","t",10,"F(t)")
plt.subplot(2,1,2)
createPlot(t3,F_kk2,"Integrationsergebnis Rechteck:","t",10,"F(t)")
```
