```python
def myIntegrate(a,b,N):  #a und b sind die Grenzen des Integrals, N die Anzahl der Blöcke
    x=np.linspace(a,b,N)
    f_x=np.sin(x)
    return sum(f_x)*(b-a)/N #Aproximation des Riemannintegrals
myIntegrate(0,np.pi,100)
    
```

```python
myIntegrate(0,2*np.pi,100) #müsste "0" sein, kommt auch annähernd hin
```

```python
import scipy.integrate as integrate
def f(x):
    return np.sin(x)
# quad benötigt eine funktion zu erst und dann die Grenzen
# die Ausgabe zeigt das Ergebnis sowie die Fehlerabweichung
print(integrate.quad(f,0,np.pi))
```

```python
n=np.linspace(0,2*np.pi,25)
f= np.cos(n)

plt.plot(f) #erst den graphen malen, dann die Beschriftung
plt.ylabel('f(n)')
plt.xlabel('n')
plt.title('cosine-wave')
plt.grid(True)

plt.stem(f)
```

```python
plt.stem(f, label="huhu")
plt.ylabel('f(n)',fontsize="28")
plt.xlabel('n',fontsize="28")
plt.title('stemps', fontsize="30", family="impact")
plt.minorticks_on() #unterstriche auf der Skala
plt.grid(True)
plt.tick_params(grid_color='green',labelsize='large')
plt.xticks(np.linspace(0,len(n),5),('0','1/2 PI','PI','3/4 PI','2PI'))
# plt.suptitle('penis')
plt.plot(2*f, label="haha")
plt.legend(loc=1)
```

```python
plt.subplot(2,1,1)
plt.plot(f)
plt.title("huhu")
plt.subplot(2,1,2)
plt.stem(2*f)
plt.title("haha")
plt.subplots_adjust(hspace=2)
plt.suptitle("hier 2 Plots")
```
