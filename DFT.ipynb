```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Satz 1 Linear
N_blk=12
n_N=np.arange(N_blk)
f_x=1/N_blk
a_1=1 # Wichtungen
a_2=2
x_1N=np.cos(2*np.pi*f_x*n_N)
x_2N=np.cos(2*np.pi*2*f_x*n_N)
x_N=a_1*x_1N+a_2*x_2N # Summer der gewichteten Signale
plt.stem(n_N,x_N)
plt.title("addierte Signale unterschiedlich gewichtet")
```

```python
X1_N=np.fft.fft(x_1N)
plt.stem(n_N,np.abs(X1_N))
plt.title("FFT von Signal 1")
```

```python
X2_N=np.fft.fft(x_2N)
plt.stem(n_N,np.abs(X2_N))
plt.stem(n_N,a_2*np.abs(X1_N))
plt.title("Spektrum von Sig2 und gewichtetem Sinal1 in einem Koordinatensystem")
```

```python
X_N=np.fft.fft(x_N)
plt.stem(n_N,np.abs(X_N))
plt.title("FFT vom summierten Signal (Sig1+Sig2)")
```

```python
#zyklische Verschiebung:
x_N=n_N
plt.stem(x_N,np.roll(n_N,2))
```
